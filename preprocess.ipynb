```python
import json

def load_json_file(file_path):
    data = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            try:
                record = json.loads(line)
                data.append(record)
            except json.JSONDecodeError:
                print(f"Skipping line due to JSONDecodeError: {line.strip()}")
    return data


file_path = '/home/cake/Documents/Sampled_Amazon.json'
data = load_json_file(file_path)

```

```python
import json

def preprocess_data(data):
    preprocessed_data = []
    for record in data:
        # Apply preprocessing here if needed
        
        # Keep only selected columns
        preprocessed_record = {key: record[key] for key in ['asin', 'brand', 'price', 'also_buy']}
        preprocessed_data.append(preprocessed_record)
    return preprocessed_data


file_path = '/home/cake/Documents/Sampled_Amazon.json'
data = load_json_file(file_path)
preprocessed_data = preprocess_data(data)

# Save preprocessed data to a JSON file
output_file_path = '/home/cake/Documents/Preprocessed_Amazon.json'
with open(output_file_path, 'w') as json_file:
    json.dump(preprocessed_data, json_file)

print("Preprocessed data saved to:", output_file_path)
```
